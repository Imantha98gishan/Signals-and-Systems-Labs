```python
import numpy as np
import matplotlib.pyplot as plt

t        = np.linspace(0,10,1000)
w        = 6
theta    = w*t
area     = 0.1
tesla    = 1

V      = area*tesla*w*np.sin(theta)

plt.figure(figsize=(10,6))
plt.plot(t,V,color='Red')
plt.title('Task 1')
plt.xlabel('time (s)')
plt.ylabel('A.B.w.sin(wt)')
plt.grid(color="black",linestyle='-',linewidth=0.25)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sampling rate
sampling_rate = 0.1

# Time range
t_discrete = np.arange(0, 10 + sampling_rate, sampling_rate)

# Calculate the sampled signal
V = 0.6 * np.sin(6 * t_discrete)

# Plotting
plt.figure(figsize=(10, 6))
plt.stem(t_discrete, V, markerfmt='o', linefmt='-', basefmt=' ', use_line_collection=True)
plt.title('Task 2')
plt.xlabel('time (s)')
plt.ylabel('A.B.w.sin(wt)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1  # Amplitude
f = 108e6  # Carrier frequency in Hz
T = 1 / f  # Period of the carrier frequency
phi = 0  # Phase angle

# Time range for 10 cycles
t = np.linspace(0, 10 * T, 1000)

# Time domain representation
x_t = A * np.sin(2 * np.pi * f * t + phi)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x_t, color='blue')
plt.title('Task 4')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Signal A
A = 1
omega = 1
phase = np.pi/3

# Writing A*cos(𝜔𝑡 + 𝜋/3) as a sum of two complex exponentials
A_exp1 = A * np.exp(1j * phase)
A_exp2 = A * np.exp(-1j * phase)

# Signal B
A1 = 2
A2 = 6
omega1 = 1
omega2 = 2
phase1 = np.pi/3
phase2 = -np.pi/4

# Writing 2*cos(𝜔𝑡 + 𝜋/3) + 6*sin(2𝜔𝑡 - 𝜋/4) as a sum of four complex exponentials
B_exp1 = A1 * np.exp(1j * phase1)
B_exp2 = A1 * np.exp(-1j * phase1)
B_exp3 = A2 * np.exp(1j * phase2)
B_exp4 = A2 * np.exp(-1j * phase2)

# Frequency range
f = np.linspace(-5, 5, 1000)

# Plotting Signal A
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(f, np.abs(A_exp1 * np.sinc(f - omega) + A_exp2 * np.sinc(f + omega)), label='Signal A')
plt.title('Two-sided frequency representation of Signal cos(wt+phi/3)')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.legend()

# Plotting Signal B
plt.subplot(2, 1, 2)
plt.plot(f, np.abs(B_exp1 * np.sinc(f - omega1) + B_exp2 * np.sinc(f + omega1) +
                    B_exp3 * np.sinc(f - omega2) + B_exp4 * np.sinc(f + omega2)), label='Signal B')
plt.title('Two-sided frequency representation of Signal 2cos(wt + phi/3) + 6sin (2wt − phi/4) ')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1.0  # Amplitude
omega = 2*np.pi  # Angular frequency (radians/sec)
phi = np.pi/3  # Phase angle

# Frequency range
f = np.linspace(0, 5, 1000)

# Single-sided frequency representation
S = A/2 * (np.sinc(f - omega/(2*np.pi)) + np.sinc(f + omega/(2*np.pi)))

# Plot
plt.figure(figsize=(8, 4))
plt.plot(f, S, label='|S(ω)|')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Single-sided frequency representation of A.cos(wt+phi)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples per cycle
N = 10

# Time axis for 5 cycles
n = np.arange(0, 5 * N)

# Square wave sampled at 10 times its frequency
x1 = np.zeros_like(n)
x1[np.mod(n, 10) > 4] = -5
x1[np.mod(n, 10) < 4] = 5

# Square wave sampled at 5 times its frequency
x2 = np.zeros_like(n)
x2[np.mod(n, 10) >= 2] = 5
x2[np.mod(n, 10) <= 2] = -5

# Half-wave rectified sinusoidal signal sampled at 8 times its frequency
x3 = np.sin(np.pi * (n+6) / 4)
x3[x3 < 0] = 0

# Full-wave rectified sinusoidal signal sampled at 4 times its frequency
x4 = np.abs(np.sin(np.pi * (n+6) / 2))

# Plotting
plt.figure(figsize=(10, 20))

plt.subplot(411)
plt.stem(n, x1)
plt.title('Square wave sampled at 10 times its frequency')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(412)
plt.stem(n, x2)
plt.title('Square wave sampled at 5 times its frequency')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(413)
plt.stem(n, x3)
plt.title('Half-wave rectified sinusoidal signal sampled at 8 times its frequency')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.subplot(414)
plt.stem(n, x4)
plt.title('Full-wave rectified sinusoidal signal sampled at 4 times its frequency')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
