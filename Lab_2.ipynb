```python
#k
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as scint

# Define the unit rectangular function
shift = 0.5

def u_rect(t):
 return np.heaviside(t-shift,0)-np.heaviside(t+shift,0)

# Define the unit triangular function
def u_tri(t):
  return scint.cumtrapz(2*(np.heaviside(t+shift,0)-2*np.heaviside(t,0)+np.heaviside(t-shift,0)),t,initial=0)

# Generate time vector
t = np.linspace(-2, 2, 1000)

# Calculate the values of the rectangular and triangular functions
u_rect_t = u_rect(t)
u_tri_t = u_tri(t)

#integral rect using cumilative trapzoidal rule
inte_rect = scint.cumtrapz(u_rect_t, t, initial=0)
#integral trian
inte_tri = scint.cumtrapz(u_tri_t, t, initial=0)

#derivative rect
deri_rect = np.gradient(u_rect_t, t)
#derivative trian
deri_tri = np.gradient(u_tri_t, t)

#Unit Rectangular
plt.figure(figsize=(10,6))
plt.plot(t,u_rect_t)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Rectangular Functions')
plt.grid(True)
plt.show()

#Unit Rectangular Integration
plt.figure(figsize=(10,6))
plt.plot(t,inte_rect)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Integral of Unit Rectangular Functions')
plt.grid(True)
plt.show()

#Unit Rectangular Diferrentiation
plt.figure(figsize=(10,6))
plt.plot(t,deri_rect)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Derivative of Unit Rectangular Functions')
plt.grid(True)
plt.show()

#Unit Triangular
plt.figure(figsize=(10,6))
plt.plot(t,u_tri_t)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Unit Triangular Functions')
plt.grid(True)
plt.show()

#Unit Triangular Integration
plt.figure(figsize=(10,6))
plt.plot(t,inte_tri)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Integral of Unit Triangular Functions')
plt.grid(True)
plt.show()

#Unit Triangular Differentiation
plt.figure(figsize=(10,6))
plt.plot(t,deri_tri)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Derivative of Unit Triangular Functions')
plt.grid(True)
plt.show()
```

```python
#b
import numpy as np


def impulse_train(t, N=10, T=20):
  dt = t[1] - t[0]
  impulse_locations = np.linspace(0, T-dt, N, dtype=int)
  impulse_train_data = np.zeros_like(t)
  impulse_train_data[impulse_locations] = 1

  if len(t) > T:
    return impulse_train_data[:int(T/dt)]
  elif len(t) < T:
    return np.pad(impulse_train_data, (0, int((T-len(t))/dt)), 'constant')

t = np.arange(0, 2.1, 0.1)

impulse_train_data = impulse_train(t)

print(impulse_train_data)
```
