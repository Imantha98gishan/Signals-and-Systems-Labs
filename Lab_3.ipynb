```python
#Task 6

import numpy as np
import matplotlib.pyplot as plt

def compute_coefficients(N):
  coefficients = []
  for n in range(1, N + 1):
    a_n = 8 / (n * 2 * np.pi) * (-1) ** ((n-1) // 2)
    coefficients.append((a_n, 0))
  return coefficients

#approx for 10 coefficient
coefficients = compute_coefficients(10)

def amplitude_single(n):
  return 2 / (np.pi * n)

def amplitude_two(n):
  return 4 / (np.pi * n)

def phase(a_n, b_n):
  return np.arctan2(b_n, a_n)

# Frequencies for plotting
freqs_single = np.arange(1, 11)
freqs_two = np.concatenate((np.arange(1, 11), -np.arange(1, 11)))

amplitudes_single = [amplitude_single(n) for n in freqs_single]
amplitudes_two = [amplitude_two(n) for n in freqs_two]
phases_single = [phase(a_n, b_n) for a_n, b_n in coefficients]
phases_two = [phase(a_n, b_n) if a_n != 0 or b_n != 0 else 0 for a_n,
b_n in coefficients]*2

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot single-sided frequency representation
axs[0, 0].stem(freqs_single, amplitudes_single, linefmt='C0-',
markerfmt='C0o', basefmt='k-')
axs[0, 0].set_title('Single-Sided Frequency')
axs[0, 0].set_xlabel('Frequency (n)')
axs[0, 0].set_ylabel('Amplitude')
axs[0, 0].grid(True)

# Plot two-sided frequency representation
axs[0, 1].stem(freqs_two, amplitudes_two, linefmt='C1-',
markerfmt='C1o', basefmt='k-')
axs[0, 1].set_title('Two-Sided Frequency')
axs[0, 1].set_xlabel('Frequency (n)')
axs[0, 1].set_ylabel('Amplitude')
axs[0, 1].grid(True)

# Plot single-side phase representation
axs[1, 0].stem(freqs_single, phases_single, linefmt='C2-',
markerfmt='C2o', basefmt='k-')
axs[1, 0].set_title('Single-Sided Phase')
axs[1, 0].set_xlabel('Frequency (n)')
axs[1, 0].set_ylabel('Phase (radians)')
axs[1, 0].grid(True)

# Plot double-sided phase representation
axs[1, 1].stem(freqs_two, phases_two, linefmt='C3-', markerfmt='C3o',
basefmt='k-')
axs[1, 1].set_title('Two-Sided Phase')
axs[1, 1].set_xlabel('Frequency (n)')
axs[1, 1].set_ylabel('Phase (radians)')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
# Show the plots
plt.show()







#Task 7

amplitude = [1, 0.5, 0.25] # Amplitude of each harmonic sinusoid

phase = [0, np.pi/3, np.pi/4] # Phase of each harmonic sinusoid

harmonic_numbers = np.arange(1, len(amplitude) + 1)

# Plot single-sided frequency (Amplitude)
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
plt.stem(harmonic_numbers, amplitude, use_line_collection=True)
plt.title('Single-Sided Frequency (Amplitude)')
plt.xlabel('Harmonic Number')
plt.ylabel('Amplitude')

# Plot single-sided frequency (Phase)
plt.subplot(2, 1, 2)
plt.stem(harmonic_numbers, phase, use_line_collection=True)
plt.title('Single-Sided Frequency (Phase)')
plt.xlabel('Harmonic Number')
plt.ylabel('Phase (radians)')

plt.tight_layout()
plt.show()


amplitude_pos = [0.5, 0.25, 0.125] # Amplitude of each harmonic

amplitude_neg = [0.5, 0.25, 0.125] # Amplitude of each harmonic

phase_pos = [0, np.pi/3, np.pi/4] # Phase of each harmonic sinusoid

phase_neg = [0, -np.pi/3, -np.pi/4] # Phase of each harmonic sinusoid

harmonic_numbers = np.arange(1, len(amplitude_pos) + 1)

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.stem(harmonic_numbers, amplitude_pos, use_line_collection=True,
basefmt=" ", linefmt='b-',
markerfmt='bo', label='Positive Frequencies')
plt.stem(-harmonic_numbers, amplitude_neg, use_line_collection=True,
basefmt=" ", linefmt='r-',
markerfmt='ro', label='Negative Frequencies')
plt.title('Two-Sided Frequency (Amplitude)')
plt.xlabel('Harmonic Number')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.stem(harmonic_numbers, phase_pos, use_line_collection=True,
basefmt=" ", linefmt='b-',
markerfmt='bo', label='Positive Frequencies')
plt.stem(-harmonic_numbers, phase_neg, use_line_collection=True,
basefmt=" ", linefmt='r-',
markerfmt='ro', label='Negative Frequencies')
plt.title('Two-Sided Frequency (Phase)')
plt.xlabel('Harmonic Number')
plt.ylabel('Phase (radians)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Task 8

#a
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 2, 1000)

T = 1.0
phi = T / 4 # Phase advanced by T/4 for 50% duty cycle

def square_wave(t, T, phi):
  return np.where(np.mod(t - phi, T) < T / 2, 1, -1)

def properly_even_square_wave(t, T, phi):
  return np.where(np.mod(t - phi, T) < T / 2, 1, -1)

plt.figure(figsize=(8, 4))
plt.plot(t, properly_even_square_wave(t, T, phi), color='blue')
plt.title(' Square Wave with 50% Duty Cycle ')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
#c
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
T = 1
omega = 2 * np.pi / T

def square_wave(t):
  return np.where(abs(t) <= T / 4, 1, -1)

def product_function(t, n):
  return square_wave(t) * np.exp(-1j * n * omega * t)

def integrate(func, a, b, n_steps):
  t = np.linspace(a, b, n_steps + 1)
  y = func(t)
  integral = np.trapz(y, t)
  return integral

def complex_fourier_coefficient(n):
  cn = integrate(lambda t: product_function(t, n), 0, T, 1000) / T
  return cn

complex_coefficients = []

for n in range(10):
  cn = complex_fourier_coefficient(n)
  complex_coefficients.append(cn)
  print(f"c{n}:", cn)

n_values = np.arange(10)
plt.stem(n_values, np.abs(complex_coefficients),
use_line_collection=True, linefmt='r-', markerfmt='r', basefmt='y-')
plt.xlabel('n')
plt.ylabel('|cn|')
plt.title('Complex Fourier Coefficients Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
s
T = 1 # Period of the square wave
num_harmonics = 10

t = np.linspace(0, T, 1000)
t_shifted = t - T/4

square_wave = np.sign(np.sin(2 * np.pi * t_shifted / T))
complex_fourier_coefficients = []
for n in range(1, num_harmonics + 1):
  c_n = np.trapz(square_wave * np.exp(-1j * 2 * np.pi * n * t_shifted/ T), t) / T
  complex_fourier_coefficients.append(c_n)

reconstructed_signal = np.zeros_like(t, dtype=complex)
for n in range(1, num_harmonics + 1):
  reconstructed_signal += complex_fourier_coefficients[n - 1] *np.exp(1j * 2 * np.pi * n * t_shifted / T)

plt.figure(figsize=(10, 6))
plt.plot(t, square_wave, label='Original Signal', color='blue')
plt.plot(t, np.real(reconstructed_signal), label='Reconstructed Signal', linestyle='--', color='red')
plt.title('Original Signal and Reconstructed Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```
