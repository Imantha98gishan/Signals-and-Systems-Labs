**Part 1: Generating and Plotting Mathematical functions**\
Most mathematical functions including Trigonometric, Exponential, Complex and various
uncommon integer functions are included in NumPy library.
*   Task 1 Generating Common Mathematical Functions as Vectors

```python
#Task 1
import numpy as np       #Mathematical Functions
import matplotlib as plt #Graphical Plotting Functions

t       = np.linspace(-10,10,100) #This define time vector for the range -10 to 10 in time step of 0.01
signal  = np.sin(t)               #Predefined numpy library

t       = np.linspace(-2,2,100)          #signal 1 time vector range -2,2 step 0.01
signal1 = 10*np.cos(6*t)                 #signal 1

t       = np.linspace(0,10,10)           #signal 2 time vector range 0,10 step 0.1
signal2 = 100*np.exp(-2*t)               #signal 2

t       = np.linspace(0,2,100)           #signal 3 time vector range 0,2 step 0.01
signal3 = 5*np.exp(-2*t)*np.cos(t)       #signal 3
```

*   Task 2 Plotting Mathematical Function Vectors

```python
#Task 2
import matplotlib.pyplot as plt    #import matlibplot library for  plotting mathematical functions

#signal plot
plt.figure(1)              #generate new graph              (a)
plt.plot(signal)           #plot signal                     (b)
plt.title('Signal')        #Title of the plot               (c)
plt.xlabel('Time (s)')     #X-axis label of the plot        (c)
plt.ylabel('sin(t)')       #Y-axis label of the plot        (c)
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid (d)

#signal 1 plot
plt.figure(2)              #generate new graph
plt.plot(signal1)          #plot signal1
plt.title('Signal 1')      #Title of the plot
plt.xlabel('Time (s)')     #X-axis label of the plot
plt.ylabel('10cos(6t)')    #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid

#signal 2 plot
plt.figure(3)              #generate new graph
plt.plot(signal2)          #plot signal2
plt.title('Signal 2')      #Title of the plot
plt.xlabel('Time (s)')     #X-axis label of the plot
plt.ylabel('100*exp(-2t)') #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid

#signal 3 plot
plt.figure(4)                   #generate new graph
plt.plot(signal3)               #plot signal2
plt.title('Signal 3')           #Title of the plot
plt.xlabel('Time (s)')          #X-axis label of the plot
plt.ylabel('5cos(t)*exp(-2t)')  #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid

#Plot of all 3 signals                                       (e)
plt.figure(5)                   #new figure
plt.plot(signal1)               #plot signal 1
plt.plot(signal2)               #plot signal 2
plt.plot(signal3)               #plot signal 3
plt.title('Signal 3')           #Title of the plot
plt.xlabel('Time (s)')          #X-axis label of the plot
plt.ylabel('Signal')            #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid

#Subplot of first 2 signals                                    (f)
plt.figure(6)
plt.subplot(211)
plt.title('Signal 1')           #subplot 1 title
plt.plot(signal1)               #plot signal 1
plt.xlabel('Time (s)')          #X-axis label of the plot
plt.ylabel('10cos(6t)')         #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid
plt.subplot(212)
plt.plot(signal2)               #plot signal 2
plt.title('Signal 2')           #subplot 2 title
plt.xlabel('Time (s)')          #X-axis label of the plot
plt.ylabel('100exp(-2t)')       #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid
```

**Part 2: Mathematical Operations on Functions using Graphical Methods**
*   Task 3: Multiplication of Two Functions



```python
#Generate two signals                                                (a)
t    = np.linspace(0,20,10)  #time vector with range 0 to 20 with step 0.1

v1   = np.sin(t)            #signal v1
v2   = np.exp(-t)           #signal v2

v1v2 = v1*v2            #product of two signals                      (b)

#All signals                                                         (c)
plt.plot(v1)
plt.plot(v2)
plt.plot(v1v2)
plt.title('Signal Multiplication')    #Title of the plot
plt.xlabel('Time (s)')                #X-axis label of the plot
plt.ylabel('Signal')                  #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25) #Grid

#Generate two functions                                              (d)
j  =  np.complex(0,1)      #defining the j
v1 =  np.exp(-t)           #signal 1
v2 =  np.exp(-j*t)         #signal 2

v1v2 = v1*v2               #product of two signal                    (e)

#Plot the complex function
plt.figure(7)                             #Generate new figure
plt.plot(np.real(v1v2),np.imag(v1v2),'k') #plot the resultant
plt.title('Signal Multiplication of Complex Function') #Title of the plot
plt.xlabel('Time (s)')                                 #X-axis label of the plot
plt.ylabel('Signal v1v2')                              #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25)  #Grid
```

*   Task 4: Derivative and Integral of a Function

```python
#Generate Signal 1
t       = np.linspace(-2,2,100)          #signal 1 time vector range -2,2 step 0.01
signal1 = 10*np.cos(6*t)                 #signal 1

df1     = np.diff(signal1)               #difference of two adjacent value in signal1      (a)
d1      = df1/0.1                        #scaled the result                                (b)

#Plotting the graphs
plt.figure(9)                            #Generate the figure
plt.plot(d1,color = "black")             #Plot the d1                                      (c)
plt.plot(signal1,color ="r")             #Plot the signal1

l1 = np.cumsum(signal1*0.1)              #Intergral function of signal 1                   (e)
plt.plot(l1,color = "blue")              #plot the function in same graph                  (f)

plt.title('Derivative Intergral of a function') #Title of the plot
plt.xlabel('Time (s)')                          #X-axis label of the plot
plt.ylabel('Signals')                           #Y-axis label of the plot
plt.grid(color="black", linestyle='-',linewidth=0.25)  #Grid
```
